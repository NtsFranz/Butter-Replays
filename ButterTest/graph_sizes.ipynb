```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = pd.read_json('butter_sizes.json', orient='index')
display(file.head())
```

```python

sns.set(style="whitegrid")
f = plt.figure(figsize=(10, 6))
f = sns.barplot(data=file[file["KeyframeInterval"] == 300], x='CompressionLevel', y='AverageFrameSizeCompressed', hue='UsingZstdDict')
plt.legend(loc='lower left', title="Zstd Dict")
```

```python
f = plt.figure(figsize=(10, 6))
f = sns.barplot(data=file[file["UsingZstdDict"] == 0], x='CompressionLevel', y='AverageFrameSizeCompressed', hue='KeyframeInterval')
```

```python
f = plt.figure(figsize=(10, 6))
f = sns.barplot(data=file[file["UsingZstdDict"] == 0], x='CompressionLevel', y='AverageFrameSizeUncompressed', hue='KeyframeInterval')
```
